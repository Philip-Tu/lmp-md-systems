## RDF Convergence Demo Notebook

This notebook demonstrates how to visually investigate the convergence of a RDF curve by applying the Central Limit Theorem.

### Imports

```python
import sys
sys.path.append('../../lib')
import pyrdf
```

First, read data of RDFs from disk using functions such as `pyrdf.read_rdf_arrays` or `pyrdf.read_multiple_files`. Ideally, the output, `rdfs` in this case, should be a 2-D numpy array.

```python
rdf_conv_file = '../data/rdf_convergence_sample.rdf'
cumul = False

_, rdfs = pyrdf.read_rdf_arrays(rdf_conv_file, cumul=cumul)
```

```python
rdfs.shape
```

To study convergence, one can check the estimated absolute error of an RDF sample against that predicted by the Central Limit Theorem and infer from that how quickly it would converge beyond an error upperbound. This can particularly be achieved with `pyrdf.plot_rdf_convergence`.

```python
Nfreq = 100000
at_bin = 68
ylim = (1e-5, 1e0)

pyrdf.plot_rdf_convergence(Nfreq, at_bin, rdfs, refline_shift=1.5, ylim=ylim)
```

**Note:** `Nfreq` should be assigned value according to what it was given in the corresponding LAMMPS input file; again, it is the frequency at which RDF is written to file. Additionally, the value of `at_bin` is to be chosen by the user depending on their bin of interest.
